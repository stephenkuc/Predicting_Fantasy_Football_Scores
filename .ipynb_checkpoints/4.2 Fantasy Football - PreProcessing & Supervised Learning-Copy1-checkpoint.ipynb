### QB Modeling

Setting up a loop to continually train a model on season data for the following year. Starting with 2019 as training data until 2021, and test data used as the following years.

### Histogram Gradient Boosting

Utilizes histograms to bin feature values; much faster speed. 

Like most GBM models, it is an ensemble learner of weak learners sequentially, trained to correct errors of previous trees.

```python
## Modeling

# setting for reproducibility

np.random.seed(11)

# getting qb relevant stats
stats = ['adp', 'age', 'avg_intended_air_yards', 'completion_percentage_above_expectation',
         'rushingatt', 'epa_total', 'te_prem_pts_pg_5ptd', 'passy_last_yr_tm_stat']
target_seasons = ['2020', '2021', '2022']

for i in range(len(target_seasons)):
    current_season = target_seasons[i]
    previous_season = str(int(current_season) - 1)
    y = 'te_prem_pts_pg_5ptd'
    
    X_train = qb[[f'{stat}_{previous_season}' for stat in stats]]
    y_train = qb[[f'te_prem_pts_pg_5ptd_{current_season}']].values.ravel()

    # Step 2: Train the Regression Models
    # Choose the regression model you want to use
    model = HistGradientBoostingRegressor()
    model.fit(X_train, y_train)

    # Step 3: Make Predictions
    X_test = qb[[f'{stat}_{current_season}' for stat in stats]]
    y_pred = model.predict(X_test)

    # Step 4: Combine Predictions
    qb[f'fantasypoints_pg_{current_season}'] = y_pred
    
    # evaluating the models
    
    y_test = qb[f'te_prem_pts_pg_5ptd_{current_season}'].values.ravel()
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"Metrics for {current_season} season:")
    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")
    print("--------------------")
    
# The wide_df DataFrame now contains the predicted data for the 2022 season for each player.
```

```python
### Is there any hyperparameter tuning we can do with these? 

# trying again, but seeing if we can include data from two seasons ago

# setting for reproducibility

np.random.seed(11)


# getting qb relevant stats
past_qb_stats = ['avg_intended_air_yards', 'ttl_opp',  'completion_percentage_above_expectation',
         'rushingatt', 'epa_total', 'te_prem_pts_pg_5ptd', 'pct_team_yards_qb']

# from 2 season ago
past_qb_stats2 = ['rushingatt', 'avg_intended_air_yards', 'te_prem_pts_pg_5ptd', 'pct_team_yards_qb']

# a feature i'd love to include, or at least explore, is current team's offense ranking last year

current_qb_stats = ['age',  'passy_last_yr_tm_stat']


target_seasons = ['2021', '2022', '2023']

for i in range(len(target_seasons)):
    current_season = target_seasons[i]
    previous_season = str(int(current_season) - 1)
    previous_season_2 = str(int(current_season) - 2) # setting for 2 seasons ago
    
    y = 'te_prem_pts_pg_5ptd'
    
        
    X_train_p = qb[[f'{stat}_{previous_season}' for stat in past_qb_stats]]
    X_train_c = qb[[f'{stat}_{current_season}' for stat in current_qb_stats]]
    
    X_train_p2 = qb[[f'{stat}_{previous_season_2}' for stat in past_qb_stats2]]
    
    column_names_p = X_train_p.columns.tolist()
    column_names_c = X_train_c.columns.tolist()
    column_names_p2 = X_train_p2.columns.tolist()
    
    
    X_train = pd.concat([X_train_p, X_train_c, X_train_p2], axis = 1, ignore_index = True)
    
    # Scaling data
    sc = StandardScaler()

    qb_trans = sc.fit_transform(X_train)

    qb_trans = pd.DataFrame(qb_trans, columns = X_train.columns)
    
    # setting up hyperparameter grid to tune over hyper parameters
    param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_iter': [100, 200, 300]}
    
    if current_season != '2023':
        y_train = qb[[f'{y}_{current_season}']].values.ravel()
        # Step 2: Train the Regression Models
        # Choose the regression model you want to use
        model = GridSearchCV(HistGradientBoostingRegressor(), param_grid, cv=5)  # You can adjust cv as needed for cross-validation
        model.fit(X_train, y_train)
        
        # Step 3: Make Predictions
        X_test_p = qb[[f'{stat}_{previous_season}' for stat in past_qb_stats]]
        X_test_p2 = qb[[f'{stat}_{previous_season_2}' for stat in past_qb_stats2]]
        X_test_c = qb[[f'{stat}_{current_season}' for stat in current_qb_stats]]
        X_test = pd.concat([X_test_p, X_test_c, X_test_p2], axis = 1, ignore_index = True)
        
        sc = StandardScaler()

        qb_trans_tst = sc.fit_transform(X_test)

        qb_trans_tst = pd.DataFrame(qb_trans_tst, columns = X_test.columns)
    
 
        y_pred = model.predict(X_test)
        # Step 4: Combine Predictions
        qb[f'fantasypoints_pg_{current_season}_hist_grad'] = y_pred
        # evaluating the models
        
        y_test = qb[f'te_prem_pts_pg_5ptd_{current_season}'].values.ravel()
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        print(f"Metrics for {current_season} season:")
        print(f"Mean Squared Error: {mse}")
        print(f"R-squared: {r2}")
        print("--------------------")
        
        print(f"Best Parameters for {current_season} season:", model.best_params_)
        
    else:
        best_model = model.best_estimator_
        predictions_23 = best_model.predict(X_train)
        qb[f'fantasypoints_pg_{current_season}_hist_grad'] = predictions_23
    X_train.columns = column_names_p + column_names_c + column_names_p2
    
    
```

## Transforming Variables

While GBM algorithms typically use decision trees as their base learners, which don't always need the most pre-processing, it is still helpful to transform skewed features. let's try doing so below:

```python
seasons = ['2019', '2020', '2021', '2022']

qb_stats = ['ttl_opp',
         'rushingatt', 'epa_total', 'te_prem_pts_pg_5ptd', 'pct_team_yards_qb', 'age']

# let's see distribution of these to transform further
for i in seasons:
    for column in qb_stats:
        plt.figure()  # Create a new figure for each histogram
        plt.hist(qb[f'{column}_{i}'], bins=20)  # Adjust the number of bins as needed
        plt.title(f'Histogram of {column} in {i}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()  # Display the histogram
```

```python
from scipy import stats

seasons = ['2019', '2020', '2021', '2022']

skew_qb_stats = ['ttl_opp',
            'rushingatt', 'epa_total', 'te_prem_pts_pg_5ptd', 'pct_team_yards_qb', 'age']

# let's see distribution of these to transform further
for i in seasons:
    for column in skew_qb_stats:

         # Apply the log transformation to make the data positive and scaled
        transformed_column = qb[f'{column}_{i}'].apply(lambda x: np.log(x - qb[f'{column}_{i}'].min() + 1))
        
        # Apply the log transformation again
        transformed_column = transformed_column.apply(lambda x: np.log(x - transformed_column.min() + 1))

        plt.figure()  # Create a new figure for each histogram
        plt.hist(transformed_column, bins=20)  # Adjust the number of bins as needed
        plt.title(f'Histogram of Transformed {column} in {i}')
        plt.xlabel(f'Transformed {column}')
        plt.ylabel('Frequency')
        
        plt.show()  # Display the histogram of the transformed data
        
        # Assign the transformed data back to the original DataFrame
        qb[f'{column}_transformed_{i}'] = transformed_column
```

```python
### Utilizing transformed data

# setting for reproducibility

np.random.seed(11)


# getting qb relevant stats
past_qb_stats = ['avg_intended_air_yards', 'ttl_opp_transformed',  'completion_percentage_above_expectation',
         'rushingatt_transformed', 'epa_total_transformed', 'te_prem_pts_pg_5ptd_transformed', 'pct_team_yards_qb_transformed']

# from 2 season ago
past_qb_stats2 = ['rushingatt_transformed', 'avg_intended_air_yards', 'te_prem_pts_pg_5ptd_transformed', 'pct_team_yards_qb_transformed']

# a feature i'd love to include, or at least explore, is current team's offense ranking last year

current_qb_stats = ['age',  'passy_last_yr_tm_stat']


target_seasons = ['2021', '2022', '2023']

for i in range(len(target_seasons)):
    current_season = target_seasons[i]
    previous_season = str(int(current_season) - 1)
    previous_season_2 = str(int(current_season) - 2) # setting for 2 seasons ago
    
    y = 'te_prem_pts_pg_5ptd'
    
        
    X_train_p = qb[[f'{stat}_{previous_season}' for stat in past_qb_stats]]
    X_train_c = qb[[f'{stat}_{current_season}' for stat in current_qb_stats]]
    
    X_train_p2 = qb[[f'{stat}_{previous_season_2}' for stat in past_qb_stats2]]
    
    column_names_p = X_train_p.columns.tolist()
    column_names_c = X_train_c.columns.tolist()
    column_names_p2 = X_train_p2.columns.tolist()
    
    
    X_train = pd.concat([X_train_p, X_train_c, X_train_p2], axis = 1, ignore_index = True)
    
    # Scaling data
    sc = StandardScaler()

    qb_trans = sc.fit_transform(X_train)

    qb_trans = pd.DataFrame(qb_trans, columns = X_train.columns)
    
    # setting up hyperparameter grid to tune over hyper parameters
    param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_iter': [100, 200, 300]}
    
    if current_season != '2023':
        y_train = qb[[f'{y}_{current_season}']].values.ravel()
        # Step 2: Train the Regression Models
        # Choose the regression model you want to use
        model = GridSearchCV(HistGradientBoostingRegressor(), param_grid, cv=5)  # You can adjust cv as needed for cross-validation
        model.fit(X_train, y_train)
        
        # Step 3: Make Predictions
        X_test_p = qb[[f'{stat}_{previous_season}' for stat in past_qb_stats]]
        X_test_p2 = qb[[f'{stat}_{previous_season_2}' for stat in past_qb_stats2]]
        X_test_c = qb[[f'{stat}_{current_season}' for stat in current_qb_stats]]
        X_test = pd.concat([X_test_p, X_test_c, X_test_p2], axis = 1, ignore_index = True)
        
        sc = StandardScaler()

        qb_trans_tst = sc.fit_transform(X_test)

        qb_trans_tst = pd.DataFrame(qb_trans_tst, columns = X_test.columns)
    
 
        y_pred = model.predict(X_test)
        # Step 4: Combine Predictions
        qb[f'fantasypoints_pg_{current_season}_hist_grad'] = y_pred
        # evaluating the models
        
        y_test = qb[f'te_prem_pts_pg_5ptd_{current_season}'].values.ravel()
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        print(f"Metrics for {current_season} season:")
        print(f"Mean Squared Error: {mse}")
        print(f"R-squared: {r2}")
        print("--------------------")
        
        print(f"Best Parameters for {current_season} season:", model.best_params_)
        
    else:
        best_model = model.best_estimator_
        predictions_23 = best_model.predict(X_train)
        qb[f'fantasypoints_pg_{current_season}_hist_grad'] = predictions_23
    X_train.columns = column_names_p + column_names_c + column_names_p2
    
    
```

### See note below on how to possibly do this

It sounds like you effectively want the geometric standard error, akin to the geometric mean exp(mean(log(x))).

While it might seem reasonable to compute that as:

exp(sd(log(x)/sqrt(n-1)))

You and others have already pointed out that that isn't correct for a few reasons. Instead, use:

exp(mean(log(x))) * (sd(log(x))/sqrt(n-1))
